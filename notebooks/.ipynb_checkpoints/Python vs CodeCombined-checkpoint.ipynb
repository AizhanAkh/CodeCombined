```python
import numpy as np
# to get plots displayed in notebook
%matplotlib notebook
import matplotlib.pyplot as plt
from math import pi
```

```python
# import all the files

# for EFT correction using python program
lcdm_pk = "/u/aizhana/Projects/PT-check/class/output/dmeff/lcdm_pk_eft.dat"               # for LCDM
dmeff0_pk = "/u/aizhana/Projects/PT-check/class/output/dmeff/dmeff_0_pk_eft.dat"               # for IDM with n = 0
dmeff2_pk = "/u/aizhana/Projects/PT-check/class/output/dmeff/dmeff_2_pk_eft.dat"             # for IDM with n = 2
dmeff_2_pk = "/u/aizhana/Projects/PT-check/class/output/dmeff/dmeff_-2_pk_eft.dat"             # for IDM with n = -2

# check LCDM model
Pk_lcdm = np.genfromtxt(lcdm_pk)[:,1]
kk_lcdm = np.genfromtxt(lcdm_pk)[:,0]
# check Dmeff model w/ n= 0
Pk_dmeff_0 = np.genfromtxt(dmeff0_pk)[:,1]
kk_dmeff_0 = np.genfromtxt(dmeff0_pk)[:,0]
# check Dmeff model w/ n= 2
Pk_dmeff_2 = np.genfromtxt(dmeff2_pk)[:,1]
kk_dmeff_2 = np.genfromtxt(dmeff2_pk)[:,0]
# check Dmeff model w/ n= -2
Pk_dmeff__2 = np.genfromtxt(dmeff_2_pk)[:,1]
kk_dmeff__2 = np.genfromtxt(dmeff_2_pk)[:,0]

# for EFT corrections using merged code

lcdm_pk_h = "/u/aizhana/Projects/CodeCombined/output/dmeff/lcdm_eft_pk_nl_pt.dat"               # for LCDM
dmeff0_pk_h = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_0_eft_pk_nl_pt.dat"               # for IDM with n = 0
dmeff2_pk_h = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_2_eft_pk_nl_pt.dat"             # for IDM with n = 2 
dmeff_2_pk_h = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_-2_eft_pk_nl_pt.dat"             # for IDM with n = -2


# check LCDM model
Pk_lcdm_h = np.genfromtxt(lcdm_pk_h)[:,1]
kk_lcdm_h = np.genfromtxt(lcdm_pk_h)[:,0]
# check Dmeff model w/ n= 0
Pk_dmeff_0_h = np.genfromtxt(dmeff0_pk_h)[:,1]
kk_dmeff_0_h = np.genfromtxt(dmeff0_pk_h)[:,0]
# check Dmeff model w/ n= 2
Pk_dmeff_2_h = np.genfromtxt(dmeff2_pk_h)[:,1]
kk_dmeff_2_h = np.genfromtxt(dmeff2_pk_h)[:,0]
# check Dmeff model w/ n= -2
Pk_dmeff__2_h = np.genfromtxt(dmeff_2_pk_h)[:,1]
kk_dmeff__2_h = np.genfromtxt(dmeff_2_pk_h)[:,0]
```

```python
# plots all Pk spectra
plt.figure()
plt.xscale('log');plt.yscale('log');plt.xlim(1.e-1, 10)
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k) \,\,\,\, [\mathrm{Mpc}/h]^3$')
plt.plot(kk_lcdm,Pk_lcdm,'k-', label = "LCDM")
plt.plot(kk_dmeff_0,Pk_dmeff_0,'g-', label = 'n = 0')
plt.plot(kk_dmeff_2,Pk_dmeff_2,'r-', label = 'n = 2')
plt.plot(kk_dmeff__2,Pk_dmeff__2,'b-', label = 'n = -2')
plt.plot(kk_lcdm_h,Pk_lcdm_h,'k--', label = "LCDM w/ Merged Code")
plt.plot(kk_dmeff_0_h,Pk_dmeff_0_h,'g--', label = 'n = 0 w/ Merged Code')
plt.plot(kk_dmeff_2_h,Pk_dmeff_2_h,'r--', label = 'n = 2 w/ Merged Code')
plt.plot(kk_dmeff__2_h,Pk_dmeff__2_h,'b--', label = 'n = -2 w/ Merged Code')
plt.legend(loc = "lower left")
#plt.title("")
plt.show()
```

```python
plt.figure()
plt.xscale('log');plt.yscale('log');plt.xlim(1.e-4, 10)
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k) \,\,\,\, [\mathrm{Mpc}/h]^3$')
plt.plot(kk_lcdm_h,(Pk_lcdm_h-Pk_lcdm)/Pk_lcdm,'k--', label = "LCDM w/ Merged Code")
plt.plot(kk_dmeff_0_h,(Pk_dmeff_0_h-Pk_dmeff_0)/Pk_dmeff_0,'g--', label = 'n = 0 w/ Merged Code')
plt.plot(kk_dmeff_2_h,(Pk_dmeff_2_h-Pk_dmeff_2)/Pk_dmeff_2,'r--', label = 'n = 2 w/ Merged Code')
plt.plot(kk_dmeff__2_h,(Pk_dmeff__2_h-Pk_dmeff__2)/Pk_dmeff__2,'b--', label = 'n = -2 w/ Merged Code')
plt.legend(loc = "lower left")
plt.title("Difference")
plt.show()
```
