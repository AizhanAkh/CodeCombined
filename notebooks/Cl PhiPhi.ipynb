```python
import numpy as np
# to get plots displayed in notebook
%matplotlib notebook
import matplotlib.pyplot as plt
from math import pi
```

```python
# Next, get all data points for models
# from output .dat files

# after "=" write path to the .dat file with Pk data (..._pk) or Cl data (..._cl)
# Output files (from .ini files above) I used are in PT-check/class/test_output/
# for example, lcdm_pk = "/Users/aizhan.akh/Documents/Projects/PT-check/class/test_output/lcdm_pk.dat"

lcdm_pk_l = "/u/aizhana/Projects/CodeCombined/output/dmeff_n0/lcdm_linear_pk.dat"               # for LCDM
lcdm_cl_l = "/u/aizhana/Projects/CodeCombined/output/dmeff_n0/lcdm_linear_cl_lensed.dat"
lcdm_pk_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/lcdm_eft_pk_nl_pt.dat"
lcdm_cl_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/lcdm_eft_cl_lensed.dat"

dmeff0_pk_l = "/u/aizhana/Projects/CodeCombined/output/dmeff_n0/dmeff_n0_linear_pk.dat"               # for IDM with n = 0
dmeff0_cl_l = "/u/aizhana/Projects/CodeCombined/output/dmeff_n0/dmeff_n0_linear_cl_lensed.dat"  
dmeff0_pk_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_0_eft_pk_nl_pt.dat" 
dmeff0_cl_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_0_eft_cl_lensed.dat" 

dmeff2_pk_l = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_2_linear_pk.dat"               # for IDM with n = 0
dmeff2_cl_l = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_2_linear_cl_lensed.dat"  
dmeff2_pk_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_2_eft_pk_nl_pt.dat" 
dmeff2_cl_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_2_eft_cl_lensed.dat" 

dmeff_2_pk_l = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_-2_linear_pk.dat"               # for IDM with n = 0
dmeff_2_cl_l = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_-2_linear_cl_lensed.dat"  
dmeff_2_pk_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_-2_eft_pk_nl_pt.dat" 
dmeff_2_cl_pt = "/u/aizhana/Projects/CodeCombined/output/dmeff/dmeff_-2_eft_cl_lensed.dat" 

# NOTE to self: CLASS outputs ClPP as ClPP*l*(l+1)!

# check LCDM model linear
Pk_lcdm_l = np.genfromtxt(lcdm_pk_l)[:,1]
kk_lcdm_l = np.genfromtxt(lcdm_pk_l)[:,0]
ll_lcdm_l = np.genfromtxt(lcdm_cl_l)[:,0]
clTT_lcdm_l = np.genfromtxt(lcdm_cl_l)[:,1]
clPP_lcdm_l = np.genfromtxt(lcdm_cl_l)[:,5]
# check LCDM model PERTURBATION THEORY
Pk_lcdm_pt = np.genfromtxt(lcdm_pk_pt)[:,1]
kk_lcdm_pt = np.genfromtxt(lcdm_pk_pt)[:,0]
ll_lcdm_pt = np.genfromtxt(lcdm_cl_pt)[:,0]
clTT_lcdm_pt = np.genfromtxt(lcdm_cl_pt)[:,1]
clPP_lcdm_pt = np.genfromtxt(lcdm_cl_pt)[:,5]


# check Dmeff model w/ n= 0 LINEAR
Pk_dmeff_0_l = np.genfromtxt(dmeff0_pk_l)[:,1]
kk_dmeff_0_l = np.genfromtxt(dmeff0_pk_l)[:,0]
ll_dmeff_0_l = np.genfromtxt(dmeff0_cl_l)[:,0]
clTT_dmeff_0_l = np.genfromtxt(dmeff0_cl_l)[:,1]
clPP_dmeff_0_l = np.genfromtxt(dmeff0_cl_l)[:,5]
# check Dmeff model w/ n= 0 PERTURBATION THEORY
Pk_dmeff_0_pt = np.genfromtxt(dmeff0_pk_pt)[:,1]
kk_dmeff_0_pt = np.genfromtxt(dmeff0_pk_pt)[:,0]
ll_dmeff_0_pt = np.genfromtxt(dmeff0_cl_pt)[:,0]
clTT_dmeff_0_pt = np.genfromtxt(dmeff0_cl_pt)[:,1]
clPP_dmeff_0_pt = np.genfromtxt(dmeff0_cl_pt)[:,5]

# check Dmeff model w/ n= 2 LINEAR
Pk_dmeff_2_l = np.genfromtxt(dmeff2_pk_l)[:,1]
kk_dmeff_2_l = np.genfromtxt(dmeff2_pk_l)[:,0]
ll_dmeff_2_l = np.genfromtxt(dmeff2_cl_l)[:,0]
clTT_dmeff_2_l = np.genfromtxt(dmeff2_cl_l)[:,1]
clPP_dmeff_2_l = np.genfromtxt(dmeff2_cl_l)[:,5]
# check Dmeff model w/ n= 2 PERTURBATION THEORY
Pk_dmeff_2_pt = np.genfromtxt(dmeff2_pk_pt)[:,1]
kk_dmeff_2_pt = np.genfromtxt(dmeff2_pk_pt)[:,0]
ll_dmeff_2_pt = np.genfromtxt(dmeff2_cl_pt)[:,0]
clTT_dmeff_2_pt = np.genfromtxt(dmeff2_cl_pt)[:,1]
clPP_dmeff_2_pt = np.genfromtxt(dmeff2_cl_pt)[:,5]

# check Dmeff model w/ n= -2 LINEAR
Pk_dmeff__2_l = np.genfromtxt(dmeff_2_pk_l)[:,1]
kk_dmeff__2_l = np.genfromtxt(dmeff_2_pk_l)[:,0]
ll_dmeff__2_l = np.genfromtxt(dmeff_2_cl_l)[:,0]
clTT_dmeff__2_l = np.genfromtxt(dmeff_2_cl_l)[:,1]
clPP_dmeff__2_l = np.genfromtxt(dmeff_2_cl_l)[:,5]
# check Dmeff model w/ n= -2 PERTURBATION THEORY
Pk_dmeff__2_pt = np.genfromtxt(dmeff_2_pk_pt)[:,1]
kk_dmeff__2_pt = np.genfromtxt(dmeff_2_pk_pt)[:,0]
ll_dmeff__2_pt = np.genfromtxt(dmeff_2_cl_pt)[:,0]
clTT_dmeff__2_pt = np.genfromtxt(dmeff_2_cl_pt)[:,1]
clPP_dmeff__2_pt = np.genfromtxt(dmeff_2_cl_pt)[:,5]
```

```python
# plots all cl/kappa/kappa spectra 

plt.figure()
plt.xscale('linear');plt.yscale('log');plt.xlim(0, 2500); #plt.ylim(1.e-2, 1.e5)
plt.xlabel(r'${\ell}$')
plt.ylabel(r'$C_{\ell} ^{\kappa\kappa}(l) $')
plt.plot(ll_lcdm_l, ((ll_lcdm_l*(ll_lcdm_l+1))**1)*clPP_lcdm_l, 'k-',label = "LCDM, Linear")
plt.plot(ll_dmeff_0_l, ((ll_dmeff_0_l*(ll_dmeff_0_l+1))**1)*clPP_dmeff_0_l, 'r-',label = "n = 0, Linear")
plt.plot(ll_dmeff_2_l, ((ll_dmeff_2_l*(ll_dmeff_2_l+1))**1)*clPP_dmeff_2_l, 'g-',label = "n = 2, Linear")
plt.plot(ll_dmeff__2_l, ((ll_dmeff__2_l*(ll_dmeff__2_l+1))**1)*clPP_dmeff__2_l, 'b-',label = "n = -2, Linear")

plt.plot(ll_lcdm_pt, ((ll_lcdm_pt*(ll_lcdm_pt+1))**1)*clPP_lcdm_pt,'k--', label = "LCDM, Perturbation Theory")
plt.plot(ll_dmeff_0_pt, ((ll_dmeff_0_pt*(ll_dmeff_0_pt+1))**1)*clPP_dmeff_0_pt,'r--', label = 'n = 0, Perturbation Theory')
plt.plot(ll_dmeff_2_pt, ((ll_dmeff_2_pt*(ll_dmeff_2_pt+1))**1)*clPP_dmeff_2_pt,'g--', label = 'n = 2, Perturbation Theory')
plt.plot(ll_dmeff_2_pt, ((ll_dmeff__2_pt*(ll_dmeff__2_pt+1))**1)*clPP_dmeff__2_pt,'b--', label = 'n = -2, Perturbation Theory')

plt.legend(loc = "lower left")
plt.title("Comparison of Cl/kappa/kappa(l) spectra of LCDM and IDM models")
plt.show()
```

```python
# plots all residual Cl/kappa/kappa spectra 

plt.figure()
plt.xscale('linear');plt.yscale('linear');plt.xlim(0, 2500); #plt.ylim(1.e-2, 1.e5)
plt.xlabel(r'${\ell}$')
plt.ylabel(r'$C_{\ell}^{\kappa\kappa}(l) $')

plt.plot(ll_dmeff_0_pt, (clPP_dmeff_0_pt - clPP_lcdm_pt)/clPP_lcdm_pt,'r--', label = 'n = 0, Perturbation Theory')
plt.plot(ll_dmeff_2_pt, (clPP_dmeff_2_pt - clPP_lcdm_pt)/clPP_lcdm_pt,'g--', label = 'n = 2, Perturbation Theory')
plt.plot(ll_dmeff__2_pt, (clPP_dmeff__2_pt - clPP_lcdm_pt)/clPP_lcdm_pt,'b--', label = 'n = -2, Perturbation Theory')

plt.legend(loc = "lower left")
plt.title("Residual Cl/kappa/kappa(l) spectra of LCDM and IDM models")
plt.show()
```
